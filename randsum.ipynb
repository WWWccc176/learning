```python
import matplotlib.pyplot as plt
import numpy as np

number = 15
x = np.linspace(-5, 5, 800)


def sumlist(N, x):

    Sum = np.zeros_like(x)
    for i in range(0, N + 1):

        Sum += np.sin((2**i) * x) / 2**i

    return Sum


a = sumlist(number, x)

sizes = np.array(np.ones_like(x)*0.1)

plt.subplot(2, 1, 1)
plt.plot(x, a, "r-")

plt.subplot(2, 1, 2)
plt.scatter(x, a, s=sizes)

plt.show()
```
